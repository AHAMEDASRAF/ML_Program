```python
import numpy as numpy
import pandas as pd
```

```python
df=pd.read_csv("/content/merged_dataset.csv")
```

```python
df.shape
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df["Year"] = df["Date"].dt.year
df["Month"] = df["Date"].dt.month
df["Day"] = df["Date"].dt.day
df["Week"] = df["Date"].dt.isocalendar().week
```

```python
df.head()
```

```python
df.drop(["Trades"],axis=1,inplace=True)
```

```python
import numpy as np
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()

# Select only numerical columns for scaling
numerical_cols = df.select_dtypes(include=['number']).columns
df_numerical = df[numerical_cols]

# Drop rows with NaN values from the numerical DataFrame
df_numerical = df_numerical.dropna()

# Check for and remove columns with zero standard deviation
# to prevent StandardScaler from producing NaNs/Infs due to division by zero.
zero_std_cols = df_numerical.columns[df_numerical.std() == 0]
if not zero_std_cols.empty:
    print(f"Warning: Dropping columns with zero standard deviation: {list(zero_std_cols)}")
    df_numerical = df_numerical.drop(columns=zero_std_cols)

scaled_data = scaler.fit_transform(df_numerical)
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)

pca_data = pca.fit_transform(scaled_data)
```
